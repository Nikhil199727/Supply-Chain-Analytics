```python
# Convert the '-' values to 0 in the 'Discount %' column and then change the data type from object to float
df_orders['Discount %'] = df_orders['Discount %'].replace('  -  ', 0).astype(float)
```

```python
import matplotlib.pyplot as plt
```

```python
# 1. Total Inventory Storage Cost
# Merging df_inventory with df_orders to calculate storage cost for each product
df_combined = pd.merge(df_orders, df_inventory, on='Product Name')
```

```python
df_combined['Storage Cost'] = df_combined['Warehouse Inventory'] * df_combined['Inventory Cost Per Unit']

# 2. Shipment Delay in Days
df_combined['Shipment Delay'] = df_combined['Shipment Day'] - df_combined['Shipment Days - Scheduled']

# 3. Profit Margin
df_combined['Profit Margin (%)'] = (df_combined['Profit'] / df_combined['Gross Sales']) * 100
```

```python
# 4. Inventory to Sales Delta
# We need to aggregate based on 'Product Name' for correct calculations
df_aggregated = df_combined.groupby('Product Name').agg({
    'Warehouse Inventory': 'sum',
    'Order Quantity': 'sum',
    'Gross Sales': 'sum',
    'Profit': 'sum'
}).reset_index()
```

```python
df_aggregated['Inventory to Sales Delta'] = df_aggregated['Warehouse Inventory'] - df_aggregated['Order Quantity']

# 5. Under or Overstock
df_aggregated['Under or Overstock'] = df_aggregated['Inventory to Sales Delta'].apply(lambda x: 'Overstock' if x > 0 else 'Understock')

# Display the DataFrame with calculated metrics
print(df_aggregated[['Product Name', 'Inventory to Sales Delta', 'Under or Overstock']])
```

```python
# 6. Plotting the Graphs

# Plot 1: Storage Cost per Product (Top 10 for better visibility)
df_storage_cost = df_combined.groupby('Product Name')['Storage Cost'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
df_storage_cost.plot(kind='bar', color='blue')
plt.title('Top 10 Products by Storage Cost')
plt.xlabel('Product')
plt.ylabel('Storage Cost')
plt.show()
```

```python
# Plot 2: Shipment Delay per Product (Top 10 for better visibility)
df_shipment_delay = df_combined.groupby('Product Name')['Shipment Delay'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
df_shipment_delay.plot(kind='bar', color='red')
plt.title('Top 10 Products by Average Shipment Delay')
plt.xlabel('Product')
plt.ylabel('Shipment Delay (Days)')
plt.show()
```

```python
# Plot 3: Profit Margin per Product (Top 10 for better visibility)
df_profit_margin = df_combined.groupby('Product Name')['Profit Margin (%)'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
df_profit_margin.plot(kind='bar', color='green')
plt.title('Top 10 Products by Profit Margin')
plt.xlabel('Product')
plt.ylabel('Profit Margin (%)')
plt.show()
```

```python
# Plot 4: Inventory to Sales Delta per Product (Top 10 for better visibility)
df_inventory_delta = df_aggregated.set_index('Product Name')['Inventory to Sales Delta'].nlargest(10)
plt.figure(figsize=(10, 6))
df_inventory_delta.plot(kind='bar', color='purple')
plt.title('Top 10 Products by Inventory to Sales Delta')
plt.xlabel('Product')
plt.ylabel('Inventory to Sales Delta')
plt.show()
```

```python
# Plot 5: Under or Overstock per Product (Top 10 for better visibility)
df_over_under_stock = df_aggregated.set_index('Product Name')['Inventory to Sales Delta'].nlargest(10)
colors = df_aggregated.set_index('Product Name')['Under or Overstock'].map({'Overstock': 'blue', 'Understock': 'orange'})
plt.figure(figsize=(10, 6))
df_over_under_stock.plot(kind='bar', color=colors)
plt.title('Top 10 Products by Inventory to Sales Delta')
plt.xlabel('Product')
plt.ylabel('Inventory to Sales Delta')
plt.show()
```

```python

import plotly.express as px
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

data = pd.read_csv("demand_inventory.csv")
print(data.head())
```

```python
data.info()
```

```python
data = data.drop(columns=['Unnamed: 0'])
```

```python
fig_demand = px.line(data, x='Date',
                     y='Demand',
                     title='Demand Over Time')
fig_demand.show()
```

```python
fig_inventory = px.line(data, x='Date',
                        y='Inventory',
                        title='Inventory Over Time')
fig_inventory.show()
```

```python

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d')





time_series = data.set_index('Date')['Demand']

differenced_series = time_series.diff().dropna()

# Plot ACF and PACF of differenced time series
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(differenced_series, ax=axes[0])
plot_pacf(differenced_series, ax=axes[1])
plt.show()
```

```python
order = (1, 1, 1)
seasonal_order = (1, 1, 1, 2) #2 because the data contains a time period of 2 months only
model = SARIMAX(time_series, order=order, seasonal_order=seasonal_order)
model_fit = model.fit(disp=False)

future_steps = 10
predictions = model_fit.predict(len(time_series), len(time_series) + future_steps - 1)
predictions = predictions.astype(int)
print(predictions)
```

```python
# Create date indices for the future predictions
future_dates = pd.date_range(start=time_series.index[-1] + pd.DateOffset(days=1), periods=future_steps, freq='D')

# Create a pandas Series with the predicted values and date indices
forecasted_demand = pd.Series(predictions, index=future_dates)

# Initial inventory level
initial_inventory = 5500

# Lead time (number of days it takes to replenish inventory)
lead_time = 1 # it's different for every business, 1 is an example

# Service level (probability of not stocking out)
service_level = 0.95 # it's different for every business, 0.95 is an example

# Calculate the optimal order quantity using the Newsvendor formula
z = np.abs(np.percentile(forecasted_demand, 100 * (1 - service_level)))
order_quantity = np.ceil(forecasted_demand.mean() + z).astype(int)

# Calculate the reorder point
reorder_point = forecasted_demand.mean() * lead_time + z

# Calculate the optimal safety stock
safety_stock = reorder_point - forecasted_demand.mean() * lead_time

# Calculate the total cost (holding cost + stockout cost)
holding_cost = 0.1  # it's different for every business, 0.1 is an example
stockout_cost = 10  # # it's different for every business, 10 is an example
total_holding_cost = holding_cost * (initial_inventory + 0.5 * order_quantity)
total_stockout_cost = stockout_cost * np.maximum(0, forecasted_demand.mean() * lead_time - initial_inventory)

# Calculate the total cost
total_cost = total_holding_cost + total_stockout_cost

print("Optimal Order Quantity:", order_quantity)
print("Reorder Point:", reorder_point)
print("Safety Stock:", safety_stock)
print("Total Cost:", total_cost)
```
